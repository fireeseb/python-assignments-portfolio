```python
def gradient_descent(x0, y0, f, grad_f, alpha, num_iterations):
    """
    Parameters:
    x0, y0: Initial point for the descent.
    f: a function of two variables
    grad_f: the gradient of f
    alpha: Learning rate.
    num_iterations: Number of iterations to perform.
    Returns:
    (x, y): The coordinates of the final point after gradient descent.
    """
    x, y = x0, y0  # Initialize x and y with the initial point

    for i in range(num_iterations):
        # obtain the gradient of f at (x, y)
        grad_x, grad_y = grad_f(x, y)
        
        # Update x and y by taking a step in the opposite direction of the gradient
        x -= alpha * grad_x
        y -= alpha * grad_y

    return x, y

def fun_1(x,y):
 return x**2+y**2
def grad_fun1(x, y):
    grad_x = 2 * x
    grad_y = 2 * y
    return grad_x, grad_y

# Test Case 1
x0, y0 = 0.1, 0.1
alpha = 0.1
num_iterations = 10
x_final, y_final = gradient_descent(x0, y0, fun_1, grad_fun1, alpha, num_iterations)
print(f"Test Case 1 Part b: Initial Point = ({x0}, {y0}), Final Point = ({x_final:.4f}, {y_final:.4f})")

# Test Case 2
x0, y0 = -1, 1
alpha = 0.01
num_iterations = 100
x_final, y_final = gradient_descent(x0, y0, fun_1, grad_fun1, alpha, num_iterations)
print(f"Test Case 2 Part b: Initial Point = ({x0}, {y0}), Final Point = ({x_final:.4f}, {y_final:.4f})")

import numpy as np

def f2(x, y):
    return 1 - np.exp(-x**2 - (y - 2)**2) - 2 * np.exp(-x**2 - (y + 2)**2)

def grad_f_2(x, y):
    grad_x = 2 * x * np.exp(-x**2 - (y - 2)**2) + 4 * x * np.exp(-x**2 - (y + 2)**2)
    grad_y = 2 * (y - 2) * np.exp(-x**2 - (y - 2)**2) + 4 * (y + 2) * np.exp(-x**2 - (y + 2)**2)
    return grad_x, grad_y

# Test Case 1
x0, y0 = 0, 1
alpha = 0.01
num_iterations = 10000
x_final, y_final = gradient_descent(x0, y0, f2, grad_f_2, alpha, num_iterations)
print(f"Test Case 1 Part c: Initial Point = ({x0}, {y0}), Final Point = ({x_final:.4f}, {y_final:.4f})")

# Test Case 2
x0, y0 = 0, -1
x_final, y_final = gradient_descent(x0, y0, f2, grad_f_2, alpha, num_iterations)
print(f"Test Case 2 Part c: Initial Point = ({x0}, {y0}), Final Point = ({x_final:.4f}, {y_final:.4f})")
print("I notice the x-value of the final point is the same as that of the initial point, which is 0 in both cases, when the gradient descent is applied from both initial points for the same maximum # of iterations and learning rate.")
print("The y-value changes by a magnitude of 1 from the initial point to the final point to a magnitude of 2, although starting from a y-value of -1 results in a decrease of 1 to a final y-value of -2 but starting from a y-value of +1 results in an increase of 1 to a final y-value of +2.")
```
