```python
import sympy as sp

x, y = sp.symbols('x y')
# Define a function
f = sp.exp(x)*sp.sin(y)+y**3
# Differentiate f with respect to x
df_dx = sp.diff(f, x)
# Differentiate f with respect to y
df_dy = sp.diff(f, y)
print (f" The derivative with respect to x is {df_dx} and the derivative with respect to y is {df_dy}")
```

```python
import sympy as sp

#defining
x, y = sp.symbols('x y')
g = x**2*y + x*y**2

#computing partial x and y derivatives
dg_dx = sp.diff(g, x)
dg_dy = sp.diff(g, y)

#finding the gradient vector
gradient_vector = (dg_dx, dg_dy)

# converting into lambda functions so we can plug in points
dg_dx_func = sp.lambdify((x, y), dg_dx, 'numpy')
dg_dy_func = sp.lambdify((x, y), dg_dy, 'numpy')

#evaluate the functions at (1, -1)
P = (1, -1)
dg_dx_evaluated = dg_dx_func(*P)
dg_dy_evaluated = dg_dy_func(*P)

#compute magnitude of the gradient vector
gradient_magnitude = (dg_dx_evaluated**2 + dg_dy_evaluated**2)**0.5


print(f" The gradient vector is {gradient_vector} and the magnitude at (1, -1) is {gradient_magnitude}")
```

```python
import sympy as sp

x, y = sp.symbols('x y')
# Define a function
h = sp.ln(x**2 + y**2)
# Differentiate f with respect to x
dh_dx = sp.diff(h, x)
# Differentiate f with respect to y
dh_dy = sp.diff(h, y)

# taking the four second partials
d2h_dx2 = sp.diff(dh_dx, x)
d2h_dy2 = sp.diff(dh_dy, y)
d2h_dxdy = sp.diff(dh_dx, y)
d2h_dydx = sp.diff(dh_dy, x)

#printing the results
print("hxx =", d2h_dx2 )
print("hyy =", d2h_dy2 )
print("hxy =", d2h_dxdy )
print("hyx =", d2h_dydx )
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#defining symbols and function
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

#making j into a recognizable function for python
j_func = sp.lambdify((x, y), j, 'numpy')

# setting plotting paramteters
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

#plotting
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#defining symbols and function
x, y = sp.symbols('x y')
k = sp.exp(-2*x**2*y**3)*sp.sin(y**3)

#making  a callable function
k_func = sp.lambdify((x, y), k, 'numpy')

#setting plotting parameters
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)


plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = e^(-2x^2 y^3)sin(y^2)')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
